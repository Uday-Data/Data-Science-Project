# Task: 
 Calculate the maximum ‘Infection Rate’ for each country and store it in a new column named ‘max_infection_rate’ 
Create a New Data Frame name ‘Corona Data’ with ‘Country/Region’ as an index and ‘max_infection_rate’ as a column 

```python
#One cannot go to each row and column to calculate the maximum infection rate for each country, in a huge data set of 
#countries upto 187 in numbers
#So we will use for loop to calculate the maximum infection rate

max_infection_rate = []

for i in corona_countrywise.index:
    max_infection_rate.append(corona_countrywise.loc[i].diff().max())
    
```

```python
max_infection_rate
```

```python
corona_countrywise['max_infection_rate'] = max_infection_rate
corona_countrywise
```

```python
#As the max_infection_rate is calculated and added as a column. Now, we will create a Data Frame named as Corona Data 
#and Country/Region as an index and max_infection_rate as a column

Corona_Data = pd.DataFrame(corona_countrywise['max_infection_rate'])
Corona_Data.head()
```

```python
Corona_Data.shape
```
